```python
import pandas as pd
table = pd.read_csv(r'C:\Users\Lucas\Downloads\supermarket_sales.csv', encoding= 'unicode_escape')
print(table)
```

```python
sale = table.groupby('City').sum()['Total']
cities = [city for city, table in table.groupby('City')]
```

```python
import plotly.express as px
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
plt.bar(cities,sale,color='pink',alpha=.6)
plt.xlabel('Cities')
plt.ylabel('Sale')
plt.title('Highest sale and city')
plt.show()
```

```python
method=table.groupby('Payment').count()['Total']
payment=['Cash','Credit card','Ewallet']
c=['m','c','r'] 
plt.figure(figsize=(12,6))
plt.bar(payment,method,color=c,alpha=0.8)
plt.xlabel('Type of Payment')
plt.ylabel('Count')
plt.title('Method of Payment')
plt.show()
```

```python
correl=table.corr()
plt.figure(figsize=(20,15))
plt.title('Correlation')
sns.heatmap(correl,annot=True, cmap="Pastel1")
plt.show()
```

```python
sale_branch=table.groupby('Branch').sum()['Total']
print('Branch C has the highest number of sale = {}$'.format(max(sale_branch)))
```

```python
plt.style.use('default')
sns.set_style('darkgrid')
branches = [branch for branch, table in table.groupby('Branch')]

plt.bar(branches,sale_branch,alpha=0.5,color='y')
plt.xlabel('Branches')
plt.ylabel('Sales')
plt.legend(['sale'])
plt.show()
```

```python
sale_product=table.groupby('Product line').sum()['Total']
products =[product for product, supermarket_df in table.groupby('Product line')]
c=['b','b','c','b','b','b']

plt.bar(products,sale_product,alpha=0.6,color=c)
plt.xticks(products, rotation='vertical')
plt.xlabel('Products')
plt.ylabel('Sales')
plt.legend(['sale'],loc='upper left')
plt.show()
```

```python
Gender_total=table.groupby('Gender').count()['Total']
gender=['Male','Female']
c=['b','r']

plt.pie(Gender_total, explode=None, labels=gender, colors=c, autopct='%1.1f%%', shadow=True, startangle=90)
plt.axis('equal')
plt.title('Gender of the members')
plt.show()
```
